```python
import math
a = np.array(range(1,11))
plt.figure(figsize=(10,6))

plt.plot(a, a**2, marker='o', c="red")
plt.plot(a, 2**a, marker='o', c="black")
plt.plot(a, np.log(a), marker='o', c="yellow")
plt.plot(a, a, marker='o', c="green")
#plt.plot(a, [math.factorial(n) for n in a], marker='o', c="pink")
plt.legend(["quadratic", "expo", "log", "linear"])
plt.title("Functions Behavior")

print(2 ** a)
```
